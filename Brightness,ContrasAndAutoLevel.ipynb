```python
#import library
import cv2
import imageio
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Masukkan gambar
img = imageio.imread("putih.jpg")
plt.imshow(img)
plt.show()
```

```python
#define reso dan tipe gambar
img_height= img.shape[0]
img_width= img.shape[1]
img_channel= img.shape[2]
img_type = img.dtype
```

```python
#membuat brightness untuk gambar grayscale
#membuat variable img_brightness untuk menampung
img_brightness = np.zeros(img.shape, dtype=np.uint8)
```

```python
#membuat fungsi penambahan brightness dengan nilai yang menjadi parameter
def brighter(nilai):
    for y  in range(0, img_height):
        for x in range(0, img_width):
            red= img[y][x][0]
            green= img[y][x][1]
            blue= img[y][x][2]
            gray= (int(red)+ int(green)+int(blue))/3
            gray += nilai
            if gray > 255:
                gray = 255
            if gray<0:
                gray = 0
            img_brightness[y][x] = (gray, gray, gray)
```

```python
#menampilkan gambar dengan parameter -100 dan +100
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) 
ax = axes.ravel() 

brighter(-100)
ax[0].imshow(img_brightness) 
ax[0].set_title("Brightness -100") 

brighter(100)
ax[1].imshow(img_brightness) 
ax[1].set_title("Brightness +100") 

fig.tight_layout
plt.show()
```

```python
#contrast
img_contrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
ddef contrass(nilai):
    # Melakukan iterasi untuk setiap piksel pada gambar
    for y in range(0, img_height):
        for x in range(0, img_width):
            # Mengambil nilai komponen warna (merah, hijau, biru) pada piksel
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            
            # Menghitung nilai keabuan (grayscale) dengan rata-rata dari komponen warna
            gray = (int(red) + int(green) + int(blue)) / 3
            
            # Menambahkan nilai kontras pada gambar
            gray += nilai
            
            # Membatasi nilai keabuan agar tidak melebihi 255 (batas maksimum)
            if gray > 255:
                gray = 255
            
            # Menetapkan nilai keabuan pada piksel gambar hasil kontras
            img_contrass[y][x] = (gray, gray, gray)
```

```python
# Membuat subplot dengan 1 baris dan 2 kolom, serta mengatur ukuran
fig, axes = plt.subplots(1, 2, figsize=(8, 8))
ax = axes.ravel()

# Mengatur kontras gambar dengan nilai 2
contrass(2)
ax[0].imshow(img_contrass)
ax[0].set_title("Contrass 2")

# Mengatur kontras gambar dengan nilai 200
contrass(200)
ax[1].imshow(img_contrass)
ax[1].set_title("Contrass 200")

# Mengatur tata letak subplot
fig.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# Membuat array kosong dengan ukuran yang sama dengan gambar
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
def autocontrass():
    # Inisialisasi nilai xmax dan xmin
    xmax = 300
    xmin = 0
    d = 0
    
    # Mendapatkan nilai d, yang akan mempengaruhi perhitungan untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            
            # Memperbarui nilai xmax dan xmin
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
    
    # Menghitung nilai d
    d = xmin - xmax
    
    # Mengatur kontras gambar secara otomatis
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            
            # Menghitung nilai keabuan baru berdasarkan tingkat kontras
            gray = int(float(255/d) * (gray - xmax))
            
            # Menetapkan nilai keabuan pada piksel gambar hasil kontras otomatis
            img_autocontrass[y][x] = (gray, gray, gray)
```

```python
# Mengatur kontras gambar secara otomatis menggunakan fungsi autocontrass()
autocontrass()

# Menampilkan gambar hasil kontras otomatis
plt.imshow(img_autocontrass)
plt.title("Contrass Autolevel")
plt.show()
```
